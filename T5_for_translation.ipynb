```python
import torch
from transformers import AutoTokenizer, AutoModelWithLMHead
```

```python
tokenizer = AutoTokenizer.from_pretrained('t5-base')
model = AutoModelWithLMHead.from_pretrained('t5-base', return_dict=True)
```

T5 can only translate to some lanuages- German, French, Romanian and 2 more.

```python
str1 = "English" #language you want to translate from
str2 = "French" #language you want to translate to
```

```python
sequence = "Hello how are you?" #text you want to translate
```

```python
str3 = "translate " + str1 + " to " + str2 + ": "
```

```python
inputs = tokenizer.encode(str3 + sequence, return_tensors = 'pt')
```

```python
outputs = model.generate(inputs)
```

```python
new_text = tokenizer.decode(outputs[0])
```

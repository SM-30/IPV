```python
import torch
from transformers import AutoTokenizer, AutoModelWithLMHead
```

T5 is available in T5 - large, base, small, 3b, 11b. 
Here I am using t5-base.

```python
tokenizer = AutoTokenizer.from_pretrained('t5-base')
model = AutoModelWithLMHead.from_pretrained('t5-base', return_dict=True)
```

Create a file with name 'text.txt' . Then copy the text that you want to summarize and paste it in this file. This file needs to be uploaded in colab, in files.
This procedure is necessary as we need the input in one single line continous line rather than multi-line. By uploading the text in a file we can easily modify it to our requirements. 

```python
f = open('text.txt', 'r')
lines = f.readlines()
lines = [line.replace("-", ",") for line in lines]
sequence = '\t'.join([line.strip() for line in lines])
```

```python
inputs = tokenizer.encode("summarize: "+ sequence, return_tensors = 'pt', max_length=512, truncation=True)
```

```python
outputs = model.generate(inputs, max_length = 150, min_length = 60, length_penalty = 5., num_beams = 2)
```

```python
summary = tokenizer.decode(outputs[0])
```

```python
print(summary)
```

```python

text_len = len(sequence)
summary_len = len(summary)
print("Length of original paragraph:", text_len, "\n")
print("Length of summary:", summary_len, "\n")
reduced = round(((text_len - summary_len)/text_len) * 100)
print("Text reduced by:", reduced, "%")
```
